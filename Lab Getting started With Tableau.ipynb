```python
import pandas as pd
import numpy as np
import datetime
import warnings

warnings.filterwarnings('ignore')
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Load the dataset into a notebook.

```python
customer_df=pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv", sep=",")
customer_df.head()
```

#### 2. Create a barplot of the number of customers per Gender.

```python
sns.set(font_scale=1.4)
customer_df['Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0)
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Number of customers", labelpad=14)
plt.title("Number of Customers per Gender", y=1.02);
```

#### 3. Create a barplot of the number of customers per EmploymentStatus and Gender.

```python
pd.crosstab(customer_df.EmploymentStatus, customer_df.Gender).plot(kind='bar')
plt.title('Number of customers per Employment Status and Gender')
plt.xlabel('Gender')
plt.ylabel( 'Customers')
```
